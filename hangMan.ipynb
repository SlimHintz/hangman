```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import nltk
import string 

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

from unidecode import unidecode 
%matplotlib inline
```

```python
# import generic list of stopwords
STOPWORDS = set(stopwords.words('english'))

# import Trump Tweets
df = pd.read_csv('data/realdonaldtrump.csv')

df.head(3)
```

We want only the "content" which contains the actual tweets themselves. We can then use some kind of vectorizer to extract the most meaningful words out of the corpus. on import we will map non-ascii characters to their ascii equivelant. 

```python
def remove_non_ascii_chars(title):
    return "".join([unidecode(char) for char in title])

# extract content and remove non-ascii characters
content = df.content.apply(remove_non_ascii_chars)
```

```python
# create a set of all unique words used in the corpus and lowercase them all 
corpus_set = set(" ".join(content.to_list()).lower().split())
```

```python
len(corpus_set)
```

So we have 98068 unique words or symbols used in the corpus. However, this includes all symbols AND all the words are currently upper case. Let's remove symbols from the corpus and also take care to lower case all of the words and then check on the size of the vocabulary after that.

```python
# Remove punctuation 
punctuation = set(string.punctuation)
corpus_no_punct = [word for word in corpus_set if not [char for char in word if char in punctuation]]
```
